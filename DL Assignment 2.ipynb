### Implement DNN for classification and Linear regression using Keras taking any dataset from Kaggle

### Classification

```python
import pandas as pd
data = pd.read_csv('diabetes.csv')
print(data)
x = data.drop("Outcome", axis=1) # all the features
y = data["Outcome"] # labels 
from keras.models import Sequential # we are using sequential model
from keras.layers import Dense
model = Sequential()
model.add(Dense(12, input_dim=8, activation="relu")) #adding a neural network with 12 neurons, input dimensions is basically 8 features
#relu short is a piecewise linear function that will output the input directly if it is positive, otherwise, it will output zero 
model.add(Dense(12, activation="relu")) #adding another layer
model.add(Dense(1, activation="sigmoid")) #one ouput layer, to check if disease is there or not
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
model.fit(x,y, epochs=150, batch_size=10) #to increase accuracy
_,accuracy = model.evaluate(x, y)
print("Model accuracy: %.2f"% (accuracy*100))
predictions = model.predict(x)
print([round(x[0]) for x in predictions])
```

### Linear Regression

```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense

# y = mx+C
data = pd.read_csv('Linear_Regression.csv')
print(data)
x = data["X"]
y = data["Y"]
model = Sequential()
model.add(Dense(150, input_dim=1, activation="relu"))
model.add(Dense(150, activation="relu"))
model.add(Dense(50, activation="relu"))
model.add(Dense(1))
model.compile(loss="mean_squared_error", optimizer="adam", metrics=["mean_squared_error"])
model.fit(x, y, epochs=200)
x = [[400]]
print(model.predict(x))
```
